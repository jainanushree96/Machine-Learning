```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=pd.read_csv("weightedX.csv")
y=pd.read_csv("weightedY.csv")
```

```python
x.head()
```

```python
x=x.values
y=y.values
```

```python
x=(x-x.mean())/x.std()
```

```python
plt.style.use("seaborn")
```

```python
plt.scatter(x,y)
plt.title("normalised")
plt.show()
```

```python
def weight(query,bw,x):
    k=(query-x)**2
    w=np.exp(-k/(2*bw*bw))
    return w
```

```python
def weightmatrix(bw,query,x):
    m=x.shape[0]
    w=np.mat(np.eye(m))
    for i in range(m):
        w[i,i]=weight(query,bw,x[i])
    return w

    
```

```python
def thetafind(x,bw,query,y):    
    w=weightmatrix(bw,query,x)
    ones=np.ones((x.shape[0],1))
    x=np.hstack((ones,x))
    first=(x.T*(w*x))
    second=x.T*(w*y)
    theta=np.linalg.pinv(first)*second
    return theta
```

```python
def hypothesis(theta,query):
    return(theta[0][0]+theta[1][0]*query)
```

```python
b=[0.01,0.1,0.3,0.8]
```

```python
for ch in b: 
    plt.figure()
    x_test=np.linspace(-2,1.5,20)
    y_test=[]
    for i in x_test:   
        theta=thetafind(x,ch,i,y)
        y_test.append(hypothesis(theta,i))
    plt.title("BW=%s"%(ch))
    plt.scatter(x_test,y_test,color='r')
    plt.scatter(np.asarray(x),np.asarray(y))
    plt.show()
```
