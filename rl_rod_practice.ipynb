```python
import numpy as np
from keras.layers import *
from keras.models import Sequential
from collections import deque
```

```python
from keras.optimizers import Adam
import random 
```

```python
class deep_q_agent:
    def __init__(self,state_size,action_size):
        self.epsilon=1
        self.memory=deque(maxlen=2000)
        self.epsilon_decay=0.995
        self.gamma=0.95
        self.epsilon_min=0.01
        self.learning_rate=0.01
        self.model=self.create_neuralmodel(state_size,action_size)
    
    def create_neuralmodel(self,state_size,action_size):
        model=Sequential()
        model.add(Dense(32,input_dim=state_size,activation="relu"))
        model.add(Dense(16,activation="relu"))
        model.add(Dense(action_size,activation="linear"))
        model.compile(loss="categorical_crossentropy",optimizer=Adam(lr=self.learning_rate))
        return model
    def remember(self,values):
        self.memory.append(values)
    def train(self,batch_size=32):
        minibatch=random.sample(self.memory,batch_size)
        for state,next_state,done,action,reward in minibatch:
            if done:
                pass
            else:
                reward=reward+self.gamma*np.amax(self.model.predict(next_state)[0])
            target_f=self.model.predict(state)
            target_f[0][action]=reward
            self.model.fit(state,target_f,epochs=1,verbose=False)
            
        return action
```

```python
agent=deep_q_agent(env.observation_space.shape[0],(env.action_space).n)
episodes=3000
batch_size=32
for e in range(episodes):
    state=env.reset()
    
    pos=abs(state[0])
    vel=abs(state[1])   
    state=np.reshape(observation,(1,len(state)))
    timesteps=2000 
    for i in range(timesteps):
        if np.random.randn(1)<=agent.epsilon:
            action=env.action_space.sample()
        else:
            action=np.argmax(agent.model.predict(state))            
        next_state,reward,done,other_info=env.step(action)
        if abs(next_state[0])>pos:
            reward=reward+1
        if abs(next_state[1])>vel:
            reward=reward+1
        if i<199 and done :
            reward=reward+10
        next_state=np.reshape(next_state,(1,len(next_state)))
        agent.remember((state,next_state,done,action,reward))

        state=next_state
        
        env.render()
        
        if done:
            print("game episode {%d}/{%d} "%(e,episodes))
            break
        
    if(len(agent.memory)>batch_size):
        agent.train()
        if agent.epsilon>agent.epsilon_min:
            agent.epsilon*=agent.epsilon_decay
    
    
env.close()
```
